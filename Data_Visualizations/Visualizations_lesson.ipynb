### Data Visualizations with Python

There are various types of plots...

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#a simple barplot with matplotlib
counties=['Nairobi','Mombasa','Machakos','Garissa','Kisumu']
schools=[105,96,140,39,160]
```

```python
plt.bar(counties,schools,color='purple',edgecolor='black')
plt.xlabel('County')
plt.ylabel('No. of schools')
plt.title('Total schools per county')
plt.show()
```

```python
plt.barh(counties,schools,color='purple',edgecolor='black')
plt.xlabel('No. of schools')
plt.ylabel('County')
plt.title('Total schools per county')
plt.show()
```

```python
plt.bar(counties,schools,color='purple',edgecolor='black')
plt.xlabel('County')
plt.ylabel('No. of schools')
plt.title('Total schools per county')

for x, y in enumerate(schools):
    plt.text(x,y+2, str(y),ha='center')
plt.show()
```

## Grouped Bar charts

```python
counties=['Nairobi','Mombasa','Machakos','Garissa','Kisumu']
boy_schools=[102,79,87,40,110]
girl_schools=[99,65,94,25,99]
```

```python
bar_width=0.35
x=np.arange(len(counties))

plt.bar(x-bar_width/2,boy_schools,bar_width,label='Boy schools',color='blue')
plt.bar(x+bar_width/2,girl_schools,bar_width,label='Girl schools',color='green')

plt.ylabel('No.of schools')
plt.xlabel('Counties')
plt.title('Comparison of Boy & Girl Schools across Counties')
plt.xticks(x,counties)
plt.legend()

plt.show()
```

## Scatter plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
# Let's generate 50 random data points for x between 0 and 10
x = np.random.uniform(0, 10, 50)

# For y, we create a relationship with some noise: y = 2*x + random noise
noise = np.random.normal(0, 2, 50)  # random noise with mean=0 and std=2
y = 2 * x + noise

# Create the scatter plot
plt.scatter(x, y, color='blue', label='Data Points', s=50)  # s sets the size of each point

# Add labels and title
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Sample Scatter Plot')

# Display a legend
plt.legend()

# Show the plot
plt.show()
```

```python
x=np.random.uniform(0,10,50)
noise=np.random.normal(0,2,50)
y=2*noise

colors=np.random.rand(50)
sizes=np.random.randint(20,300,size=50)

plt.figure(figsize=(10,6))
scatter=plt.scatter(x,y,c=colors,s=sizes,alpha=0.7,cmap='viridis')

#add a color bar
cbar=plt.colorbar(scatter)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Seed for reproducibility
np.random.seed(0)

# Generate sample data for 50 cities:
# x and y represent geographic coordinates (e.g., longitude and latitude)
x = np.random.uniform(-100, -70, 50)   # Example longitudes
y = np.random.uniform(30, 50, 50)       # Example latitudes

# Generate temperature values (in C) for each city (ranging from 0C to 40C)
temperatures = np.random.uniform(0, 40, 50)

plt.figure(figsize=(10, 6))

# Create a scatter plot where the color of each point is based on the temperature
scatter = plt.scatter(x, y, c=temperatures, cmap='coolwarm', s=100, edgecolor='black')

# Add a color bar that shows the mapping between colors and temperature values
cbar = plt.colorbar(scatter)
cbar.set_label('Temperature (C)')

# Add labels and a title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('City Temperatures')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Seed for reproducibility
np.random.seed(0)

# Generate sample data for 50 cities:
x = np.random.uniform(-100, -70, 50)   # Example longitudes
y = np.random.uniform(30, 50, 50)       # Example latitudes
temperatures = np.random.uniform(0, 40, 50)  # Temperature values in C

plt.figure(figsize=(10, 6))

# Create the scatter plot, coloring points by temperature
scatter = plt.scatter(x, y, c=temperatures, cmap='coolwarm', s=100, edgecolor='black')
cbar = plt.colorbar(scatter)
cbar.set_label('Temperature (C)')

# Let's annotate the city with the highest temperature
max_temp_index = np.argmax(temperatures)  # Find index of max temperature

# Use plt.annotate() to add a text annotation with an arrow pointing to the data point
plt.annotate(
    f"Max Temp: {temperatures[max_temp_index]:.1f}C",  # Text to display
    (x[max_temp_index], y[max_temp_index]),             # The point to annotate
    textcoords="offset points",                         # Use offset coordinates for the text
    xytext=(10, 10),                                    # Offset position of the text
    arrowprops=dict(arrowstyle="->", color='black')      # Style of the arrow
)

# Add labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('City Temperatures with Annotation')
plt.show()
```
