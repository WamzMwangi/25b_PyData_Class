```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Plotting contour plots

```python
#load the data
nai_temp=pd.read_csv(r'C:\Users\Pula Advisors\Desktop\25b_PyData_Class\datasets\Artificial_Temperature_Dataset.csv')
```

```python
nai_temp.head()
```

```python
from scipy.interpolate import griddata
#extract the data
X=nai_temp['Longitude'].values
Y=nai_temp['Latitude'].values
Z=nai_temp['Temperature'].values

#create a grid for interpolation
xi = np.linspace(min(X), max(X), 100)
yi = np.linspace(min(Y), max(Y), 100)
Xi, Yi = np.meshgrid(xi, yi)
Zi = griddata((X, Y), Z, (Xi, Yi), method='cubic')

# Step 4: Create contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(Xi, Yi, Zi, levels=50, cmap='coolwarm')


# Labels and title
plt.title("Temperature Distribution in a City")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

### Plotting Histograms & KDE Plots

```python
#load data
earthquake_df=pd.read_csv(r'C:\Users\Pula Advisors\Desktop\25b_PyData_Class\datasets\Earthquakes_v3.csv')
earthquake_df.head()
```

```python
earthquake_df.shape
```

```python
#plot the histogram
plt.figure(figsize=(10,6))
plt.hist(earthquake_df['MAGNITUDE'],bins=20,label='Magnitude',color='blue',alpha=0.5)
plt.xlim(1,4)
plt.ylim(0,40000)
plt.xticks(range(0,4,1))
plt.show()
```

```python
modified_df=earthquake_df.loc[0:2000]
modified_df.shape
```

```python
plt.figure(figsize=(10,6))

sns.kdeplot(data=earthquake_df,x='MAGNITUDE',fill=True,vertical=True)


plt.show()
```

```python
#plotting a kde plot
plt.figure(figsize=(10,6))

sns.kdeplot(data=modified_df,x='MAGNITUDE',y='DEPTH',fill=True,cbar=True)
plt.ylim(0,20)

plt.show()
```

## Plotting Boxplots

```python
data=earthquake_df.loc[earthquake_df['MAGNITUDE']<2]
data.head()
```

```python
#plot boxplots

plt.figure(figsize=(10,6))

sns.violinplot(data=data,x='MAGNITUDE',y='DEPTH')

plt.show()
```

```python
# Generate two groups: City A and City B temperatures
city_A = np.random.normal(loc=27, scale=3, size=1000)  # Warmer climate
city_B = np.random.normal(loc=22, scale=4, size=1000)  # Cooler climate

# Plot both histograms
plt.figure(figsize=(8, 6))
plt.hist(city_A, bins=20, alpha=0.5, label="City A (Warmer)", color="red", edgecolor="black")
plt.hist(city_B, bins=20, alpha=0.5, label="City B (Cooler)", color="blue", edgecolor="black")

# Labels, title, and legend
plt.title("Temperature Comparison Between Two Cities")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```
