# Aggregation & Grouping

```python
data.head()
```

```python
data.groupby(['Memory','Storage'])['Selling Price'].mean()
```

```python
data.pivot_table()
```

```python
import pandas as pd

# Create a pivot table
pivot_avg_price = pd.pivot_table(
    data,
    values='Selling Price',
    index='Mobile_Brand',  # Rows: each brand
    columns='Colors',           # Columns: different colours
    aggfunc='mean',             # Calculate the average
    fill_value=0                # Replace NaNs with 0 if no data is available
)

pivot_avg_price
```

```python
data['Selling Price'].mean()
```

```python
#which brand has been given the highest discount
data.groupby('Mobile_Brand')['Discount'].mean().sort_values(ascending=False)
```

```python
data.groupby('Mobile_Brand')['Models'].nunique().sort_values(ascending=False)
```

```python
data.groupby('Mobile_Brand')['Models'].nunique()
```

```python
# Group by mobile_brand_name and calculate the selling price range for each brand
price_range = data.groupby('Mobile_Brand')['Selling Price'].agg(lambda x: x.max() - x.min())

# Print the result
print(price_range)
```

```python
data.groupby('Mobile_Brand')[['Memory','Storage']].max()
```

```python
data.groupby('Mobile_Brand')[['Selling Price','Discount']].mean()
```
