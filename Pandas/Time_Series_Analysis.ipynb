## Time Series Analysis
1. Converting objects to Datetime objects
2. Extracting other time data from Datetime objects
3. Resampling df using datetime objects (Groupby of timeseries)
4. Rolling windows

```python
air_data=pd.read_csv(r'C:\Users\Pula Advisors\Desktop\PyData_Class\datasets\air_quality_no2_long.csv')
air_data.head()
```

```python
air_data.info()
```

```python
air_data.loc[0,'date.utc'].day_name()
```

```python
air_data['date.utc']=pd.to_datetime(air_data['date.utc'],format='mixed')
air_data.head()
```

```python
air_data['date.utc'].dtypes
```

```python
air_data.loc[0,'date.utc'].day_name()
```

```python
#extracting year, month, day from datetime object
air_data['Year']=air_data['date.utc'].dt.year
air_data['Month']=air_data['date.utc'].dt.month
air_data['Day']=air_data['date.utc'].dt.day

air_data.head()
```

```python
air_data['Year'].nunique()
```

```python
air_data['Hour']=air_data['date.utc'].dt.hour
air_data.head()
```

```python
#resampling method in time series data
hourly_df=air_data['date.utc'].asfreq('h')
```

```python
hourly_df
```

```python
air_data=air_data.set_index('date.utc')
air_data
```

```python
paris_df=air_data.loc[air_data['city']=='Paris']
paris_df.head()
```

```python
sec_df=paris_df.asfreq(freq='30S',method='ffill')
sec_df
```

```python
#resampling
hourly_df=air_data.resample('m').max(numeric_only=True)
hourly_df
```

```python
#Rolling windows in timeseries data
paris_df['3_day_pollution']=paris_df['value'].rolling(window=3).sum()
```

```python
paris_df.head()
```

```python
paris_df['3_day_pollution']=paris_df['value'].rolling(window=3,min_periods=1).sum()
paris_df.head()
```
